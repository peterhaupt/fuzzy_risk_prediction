```python
import pandas as pd

# Load the train and test DataFrames
train_df = pd.read_pickle("data/train.pkl")
test_df = pd.read_pickle("data/test.pkl")

# Display the first few rows of the DataFrames
print("Train DataFrame:")
display(train_df.head())

print("Test DataFrame:")
display(test_df.head())

# Check correlations for numerical columns in the train DataFrame
print("\nCorrelation Matrix (Train DataFrame):")
train_corr = train_df.select_dtypes(include=['int64', 'float64']).corr()
display(train_corr)

# Check correlations for numerical columns in the test DataFrame
print("\nCorrelation Matrix (Test DataFrame):")
test_corr = test_df.select_dtypes(include=['int64', 'float64']).corr()
display(test_corr)

# Save the correlation matrices as CSV files (optional)
train_corr.to_csv("data/train_correlations.csv")
test_corr.to_csv("data/test_correlations.csv")
```
